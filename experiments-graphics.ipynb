```python
import pandas as pd
import matplotlib.pyplot as plt

path_dt = './experiments/accuracy_results_dt.csv'  
path_svm = './experiments/accuracy_results_svm.csv'
path_lr = './experiments/accuracy_results_lr.csv'

data_dt = pd.read_csv(path_dt).head(10)
data_svm = pd.read_csv(path_svm).head(10)
data_lr = pd.read_csv(path_lr).head(10)

data_lr['accuracy'] = data_lr['accuracy'] * 100

# Configuracin de la grfica de Accuracy vs. n_components
plt.figure(figsize=(12, 6))
plt.plot(data_dt['n_components'], data_dt['accuracy'], label='Decision Tree', marker='o')
plt.plot(data_svm['n_components'], data_svm['accuracy'], label='SVM', marker='o')
plt.plot(data_lr['n_components'], data_lr['accuracy'], label='Logistic Regression', marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy vs. Number of Components')
plt.legend()
plt.grid(True)
plt.show()

# Configuracin de la grfica de Time vs. n_components (Decision Tree && SVM && Logistic Regression)
plt.figure(figsize=(12, 6))
plt.plot(data_dt['n_components'], data_dt['time'], label='Decision Tree', marker='o')
plt.plot(data_svm['n_components'], data_svm['time'], label='SVM', marker='o')
plt.plot(data_lr['n_components'], data_lr['time'], label='Logistic Regression', marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Time (seconds)')
plt.title('Time vs. Number of Components')
plt.legend()
plt.grid(True)
plt.show()



# Configuracin de la grfica de Time vs. n_components (SVM && Logistic Regression)
plt.figure(figsize=(12, 6))
plt.plot(data_svm['n_components'], data_svm['time'], label='SVM', marker='o')
plt.plot(data_lr['n_components'], data_lr['time'], label='Logistic Regression', marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Time (seconds)')
plt.title('Time vs. Number of Components')
plt.legend()
plt.grid(True)
plt.show()
```
